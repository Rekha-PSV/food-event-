**Step 5 : Data visualization**

```python
import matplotlib.pyplot as plt
```

```python
data.columns
```

**Harzontal Bar Graph**

```python
# Assuming 'data' is your DataFrame
grouped = data.groupby('PRI_FDA Industry Name')['CI_Age at Adverse Event'].mean().head(10)
grouped.plot(kind='barh')
plt.title('CI_Age at Adverse Event by PRI_FDA Industry Name')
plt.xlabel('CI_Age at Adverse Event')
plt.ylabel('PRI_FDA Industry Name')
plt.show()
```

**pie chart**

```python
# Assuming 'data' is your DataFrame
grouped = data.groupby('PRI_Product Role').size()

# Plotting the pie chart
grouped.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('CI_Gender')
plt.ylabel('')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

**Line chart**

```python
# Assuming 'data' is your DataFrame
grouped = data.groupby('AEC_One Row Outcomes')['CI_Age at Adverse Event'].mean().head(20)
grouped.plot(kind='line', marker='o')
plt.title('CI_Age at Adverse Event by AEC_One Row Outcomes')
plt.xlabel('AEC_One Row Outcomes')
plt.ylabel('CI_Age at Adverse Event')
plt.xticks(rotation=45)
plt.show()
```

**Scatter Graph**

```python
 # Assuming 'data' is your DataFrame
grouped = data.groupby('PRI_Product Role')
for name, group in grouped:
    plt.scatter(group['CI_Gender'], group['CI_Age at Adverse Event'], label=name, alpha=0.5)
plt.title('CI_Age at Adverse Event vs CI_Gender by PRI_Product Role')
plt.xlabel('CI_Gender')
plt.ylabel('CI_Age at Adverse Event')
plt.legend(title='CI_Gender by PRI_Product Role')
plt.show()
```

**Bar Graph**

```python
# Assuming 'data' is your DataFrame
grouped = data.groupby('PRI_Reported Brand/Product Name')['CI_Age at Adverse Event'].mean().head(10)
grouped.plot(kind='bar')
plt.title('Mean CI_Age at Adverse Event by PRI_Reported Brand/Product Name')
plt.xlabel('PRI_Reported Brand/Product Name')
plt.ylabel('Mean CI_Age at Adverse Event')
plt.show()
```

**Step 6 : Aggregate Function**

```python
# groupby function for Customer ID and Quantity
df = pd.DataFrame(data)
grouped=df.groupby('PRI_FDA Industry Name')['CI_Age at Adverse Event'].count().head(20)
print(grouped)
```

```python
# Assuming data is your DataFrame
aggregate_functions = {
    'PRI_FDA Industry Code': ['sum', 'mean', 'median', 'min', 'max', 'std', 'count'],
}
summary_statistics = data.agg(aggregate_functions)
print(summary_statistics)
```
